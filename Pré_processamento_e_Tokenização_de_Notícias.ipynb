#2. Pré-processamento do Texto

```python
# Bibliotecas
import re
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
!pip install spacy
!python -m spacy download pt_core_news_sm
import spacy
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# Guardando os textos em uma variável e separando-os
textos = noticias['text']
# print(textos)
for texto in range(len(textos)):
  if isinstance(textos.iloc[texto], str):
    texto_separado = textos.iloc[texto].split()
    # print(texto_separado)
```

```python
# Removendo caracteres especiais, números e pontuações
novo_texto = []
for palavra in texto_separado:
  if isinstance(palavra, str):
    palavra = re.sub(r'[,.!?]', '', palavra)
    palavra = re.sub(r'[0-9]', '', palavra)
    novo_texto.append(palavra)

# print(novo_texto)
```

```python
# Padronizando o texto em minúsculas
novo_texto_minusculo = []
for palavra in novo_texto:
  if isinstance(palavra, str):
    palavra = palavra.lower()
    novo_texto_minusculo.append(palavra)

# print(novo_texto_minusculo)
```

```python
# Lemmatização
nlp = spacy.load('pt_core_news_sm')
data = []

for palavra in novo_texto_minusculo:
  if isinstance(palavra, str):
    doc = nlp(palavra)
    for token in doc:
      data.append({
        "Texto": token.text,
        "Lemma": token.lemma_,
        "POS": token.pos_,
        "Dependência": token.dep_
        })

df = pd.DataFrame(data)

print(df)
```

```python
# Remoção de stop words
stop_words = stopwords.words('portuguese')

with open("stopwords.txt", "w") as output:
  for word in stop_words:
      output.write(word+'\n')

for palavra in novo_texto_minusculo:
  if isinstance(palavra, str):
    tokens = word_tokenize(palavra)
  palavras_sem_stopwords  = [token for token in tokens if token not in stop_words]
  print(palavras_sem_stopwords)
```

#3. Técnicas de Tokenização

```python
# Tokenização Simples
novo_texto_minusculo_tokenizado = []
for palavra in novo_texto_minusculo:
  if isinstance(palavra, str):
    tokens = word_tokenize(palavra)
    novo_texto_minusculo_tokenizado.append(tokens)

print(novo_texto_minusculo_tokenizado)
```

```python
# Tokenização por N-gramas
# Bibliotecas
from nltk.lm.preprocessing import padded_everygram_pipeline, flatten
from nltk.lm import MLE
from nltk.lm.preprocessing import pad_both_ends
from nltk.util import everygrams
from nltk.lm import Vocabulary

# Aplicação
texto_tok_ngramas = []
for palavra in novo_texto_minusculo_tokenizado:
  if isinstance(palavra, str):
    tokens = word_tokenize(palavra)
    texto_tok_ngramas.append(tokens)
bigramas = 2
ngramas_pad, vocab = padded_everygram_pipeline(bigramas, texto_tok_ngramas)

lm = MLE(bigramas)
lm.fit(ngramas_pad, vocab)

trigramas = 3
ngramas_pad, vocab = padded_everygram_pipeline(trigramas, texto_tok_ngramas)

lm = MLE(trigramas)
lm.fit(ngramas_pad, vocab)
```
